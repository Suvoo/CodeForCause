```python
from PIL import Image
```

```python
img = Image.open("thor.jpg")
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
arr = np.array(img)
arr.shape # 3 represents rgb layer
```

```python
plt.imshow(arr)
```

```python
face = arr[15:110,60:160] #column then row (y then x)
```

```python
plt.imshow(face)
```

```python
import cv2
```

```python
classifier = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
faces = classifier.detectMultiScale(arr) #to find all available faces
faces
```

```python
face = faces[0]
face
```

```python
x,y,w,h = face
chopped = arr[y:y+h,x:x+w]
```

```python
plt.imshow(chopped)
```

```python
gen = cv2.rectangle(arr, (x,y), (x+w,y+h), (255,0,0),3) 
#to draw a rectangle around the face
#rectangle( image , start point, end points, color, thickness)
```

```python
plt.imshow(gen)
```
