```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

import cv2
```

```python
frame = cv2.imread("thor.jpg")
plt.imshow(frame) #openCV reads in BGR, so REVERSE
```

```python
rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
plt.imshow(rgb)
```

```python
classifier = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
faces = classifier.detectMultiScale(frame)
faces #to detect all faces
```

```python
face = faces[0]
face
```

```python
x, y, w, h = face
chop = rgb[y:y+h, x:x+w]
```

```python
plt.imshow(chop)
```

```python
gen = cv2.rectangle(rgb, (x,y), (x+w,y+h), (255,0,0),10) 
#to draw a rectangle around the face
#rectangle( image , start point, end points, color, thickness)
```

```python
plt.imshow(gen)
```
