```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
pil = Image.open("thor.jpg")
pil
```

```python
arr = np.array(pil)
arr.shape # rows,cols,dept(RGB)
```

```python
cut = arr[:5,:5] #to show only first 5x5 pixels
plt.imshow(cut)
```

```python
plt.imshow(arr)
```

```python
work = arr.copy()
```

```python
work[:,:,[1,2]] = 0 #putting green and blue as 0, so only red
plt.imshow(work)
```

```python
work[:,:,[1,2]] = 123 #putting green and blue as any b/w 0 255 so only red
plt.imshow(work)
```

```python
gray = arr.mean(axis=2) #mean of all colors on axis 2
plt.imshow(gray) #uses multicolor grayscale
```

```python
plt.imshow(gray, cmap="gray") #colormaps
```
