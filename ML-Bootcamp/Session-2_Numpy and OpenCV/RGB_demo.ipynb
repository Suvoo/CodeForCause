```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
pil = Image.open("RGB_Matte_Hierarchy.png")
pil
```

```python
arr = np.array(pil)
arr.shape # rows,cols,dept(RGB)
```

```python
cut = arr[:5,:5] #to show only first 5x5 pixels
plt.imshow(cut)
```

```python
red = arr[:,:,0] #only want depth 0 thats red
plt.imshow(red) #only shows red part
```

```python
green = arr[:,:,1] #only want depth 1 thats green
plt.imshow(green) #only shows green part
```

```python
red1 = arr[200:300,200:400,0] #only want depth 0 thats red
plt.imshow(red1) #only shows red part slicedd
```
